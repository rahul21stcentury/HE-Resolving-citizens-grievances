```python
def process(data):
    data=data.drop(['sharepointid','application','country.alpha2','decisiondate','docname','itemid','judgementdate','kpdate','languageisocode','originatingbody_type','parties.0','parties.1','parties.2','respondent.0','respondent.1','respondent.2','respondent.3',
           'respondent.4'],axis=1)
    data['ecli']=data['ecli'].iloc[:].str.slice(13,17)
    colnames=data.columns
    values=data.isna().sum()
    values
    empty=[]
    for i in range(len(colnames)):
        if(values[i]!=0):
            print(colnames[i]+"       "+str(values[i]))
            empty.append(colnames[i])

    data=data.drop(empty,axis=1)
    data.head(2)

    #data.to_csv("AS.csv")

    data['ecli']=pd.to_numeric(data['ecli'])

    le = preprocessing.LabelEncoder()

    data['doctypebranch']=le.fit_transform(data['doctypebranch'])

    data['separateopinion']=le.fit_transform(data['separateopinion'])

    data['ecli']-=1968

    data.head(3)

    e1=pd.get_dummies(data['country.name'])
    e2=pd.get_dummies(data['originatingbody_name'])
    e3=pd.get_dummies(data['originatingbody'],prefix='body_')
    e3=pd.get_dummies(data['typedescription'],prefix='type')

    df=pd.concat([e1,e2,e3],axis=1)
    df

    data=data.drop(['country.name','originatingbody_name','originatingbody','typedescription','respondentOrderEng'],axis=1)
    data=pd.concat([data,df],axis=1)

    appno=data['appno']
    appno
    return data
```

```python
data_1=process(data)
data_1
```

```python
data_2=process(data_test)
data_2
```
